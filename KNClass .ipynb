```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score

# Splitting dataset into training and testing sets
XHV_train, XHV_test, yHV_train, yHV_test = train_test_split(XHV, yHV, test_size=0.2, random_state=42)

# Scaling the features
scaler = StandardScaler()
XHV_train_scaled = scaler.fit_transform(XHV_train)
XHV_test_scaled = scaler.transform(XHV_test)

# Applying SMOTE on the scaled training data
smote = SMOTE(random_state=42)
XHV_train_smote, yHV_train_smote = smote.fit_resample(XHV_train_scaled, yHV_train)

# Training the k-NN classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(XHV_train_smote, yHV_train_smote)

# Making predictions on the test set
yHV_pred = knn.predict(XHV_test_scaled)
print("Accuracy:", accuracy_score(yHV_test, yHV_pred))
print(classification_report(yHV_test, yHV_pred))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(XHV_train, yHV_train)
```

```python
yHV_pred = model.predict(XHV_test)


mse = mean_squared_error(yHV_test, yHV_pred)
r2 = r2_score(yHV_test, yHV_pred)

print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```
